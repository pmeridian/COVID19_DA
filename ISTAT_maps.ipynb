```python
import geopandas as gpd
comuni = gpd.read_file('municipalities.geojson')
```

```python
comuni.head(5)
```

```python
import pandas as pd
urlIstat = 'dataISTAT.json'
dfIstat = pd.read_json(urlIstat,orient='index')
```

```python
dfIstat.head(5)
```

```python
merged=comuni.merge(dfIstat, left_on = 'PRO_COM', right_on = 'COD_PROVCOM',how = 'left')
```

```python
merged.head(10)
```

```python
import matplotlib.animation as animation
f1, ax1 = plt.subplots(1,figsize=(18, 14))

txt = plt.text(0.35, 0.9,
               'Excess mortality pro capite Italy',
               fontsize = 16, transform=f1.transFigure)
txt1 = plt.text(0.24, 0.15,
               'Source ISTAT',
               fontsize = 10, transform=f1.transFigure)
txt2 = plt.text(0.41, 0.88,
             'Week 6',
             fontsize = 14, transform=f1.transFigure)
def init():
    ax1.set_axis_off()
    plt.axis('equal')
    comuni.boundary.plot(ax=ax1,linewidth=0.2, edgecolor='0.5')

def animate(i):
    txt2.set_text('Week %d'%i)
    leg=True if i==6 else False
    fr=merged.dropna().plot(ax=ax1, linewidth=0., edgecolor='0.5',column='ECCESSO_PROCAPITE_SETTIMANA_%d'%i,
                    vmin=0,vmax=0.001,
                    legend=leg, cmap='Reds'
                    )

anim = animation.FuncAnimation(f1, animate, interval=1000, init_func=init, frames=range(6,14), repeat = False)
anim.save('/Users/meridian/cernbox/www/plots/COVID19_DA_ISTAT/mortality.mp4')
```
